```python
import mysql.connector
import time

mysql_query_times["popular_featured_track_query"] = []

mydb = mysql.connector.connect(
  host="localhost",
  user="root",
  password="alex1234",
  database="spotify"
)

mycursor = mydb.cursor()

# Select the arists name and their most popular featured track,the album it belongs to and  it's popularity
# order by track popularity in descending order
query= """
SELECT a1.name, t1.name, al1.name, t1.popularity 
  FROM artists a1 
  INNER JOIN track_artists ta1 ON a1.id = ta1.artist_id 
  INNER JOIN tracks t1 ON ta1.track_id = t1.id 
  INNER JOIN albums al1 ON t1.album_id = al1.id 
  WHERE t1.id IN (
    SELECT ta2.track_id 
    FROM track_artists ta2
    GROUP BY ta2.track_id
    HAVING COUNT(*) > 1
    ) AND 
    t1.popularity = (
    SELECT MAX(t2.popularity) 
    FROM tracks t2
    INNER JOIN track_artists ta2 ON t2.id = ta2.track_id
    WHERE  ta2.artist_id = a1.id AND t2.id IN (
      SELECT ta3.track_id
      FROM track_artists ta3
      GROUP BY ta3.track_id
      HAVING COUNT(*) > 1
      )
    )
  GROUP BY a1.id, t1.id
  ORDER BY t1.popularity DESC
"""
# result = mycursor.execute(query)
# myresult = mycursor.fetchall()
# print(" artist_name,track_name,album_name, track_popularity")
# for x in myresult:
#   print(x)

for i in range(30):
  start_time = time.time()
  mycursor.execute(query)
  mysql_query_times["popular_featured_track_query"].append(time.time() - start_time)
  mycursor.fetchall()

```

### Insert

```python
import mysql.connector
import time

mydb = mysql.connector.connect(
  host="localhost",
  user="root",
  password="alex1234",
  database="spotify"
)

mysql_query_times["insert"] = []
mycursor = mydb.cursor()


for i in range(30):
  id = i+200000
  start = time.time()
  # insert an artist with name "inserted_artist"
  query = f"""
  INSERT INTO artists (id,name) VALUES ({id},'inserted_artist{i}')
  """
  mycursor.execute(query)

  #insert an album of this artist with name "inserted_album"
  query = f"""
  INSERT INTO albums (id,name, artist_id) VALUES ({id},'inserted_album{i}', (SELECT id FROM artists WHERE name = 'inserted_artist{i}'))
  """
  mycursor.execute(query)

  #insert a track of this album with name "inserted_track" and popularity 100, duration_ms 5000, explicit 1, loudness -5 
  query = f"""
  INSERT INTO tracks (id,name, album_id, popularity, duration_ms, explicit, loudness) 
  VALUES ({id},'inserted_track{i}', (SELECT id FROM albums WHERE name = 'inserted_album{i}'), 100, 5000, 1, -5)
  """
  mycursor.execute(query)

  #insert in track artists the relation between the inserted track and the inserted artist
  query = """
  INSERT INTO track_artists (track_id, artist_id) 
  VALUES ((SELECT id FROM tracks WHERE name = 'inserted_track{i}'), (SELECT id FROM artists WHERE name = 'inserted_artist{i}'))
  """
  mycursor.execute(query)
  mysql_query_times["insert"].append(time.time() - start)
  

mydb.commit()

```

```python
import pandas as pd

my_df = pd.DataFrame.from_dict(mysql_query_times)
my_df.to_csv('mysql_with_optimizations_times.csv', index=False, header=True)
```

```python
from pymongo import MongoClient
import pprint
import time

mongo_query_times["popular_featured_track_query"] = []

client = MongoClient('localhost', 27017)
db = client.spotify

artists = db.artists
albums = db.albums
tracks = db.tracks

# query the arists name and their most popular featured track,the album it belongs to and  it's popularity
# order by track popularity
pipeline = [
    {"$unwind": "$tracks"},
    {"$match": {"$expr": {"$gt": [{"$size": "$tracks.artists"}, 1]}}},
    {"$unwind": "$tracks.artists"},
    # {"$lookup": {
    #     "from": "artists",
    #     "localField": "tracks.artists._id",
    #     "foreignField": "_id",
    #     "as": "artist_obj"
    # }},
    {"$unwind": "$tracks.artists"},
    {"$sort": {"tracks.popularity": -1}},
    {"$group": {
        "_id": "$tracks.artists.name",
        "track_name": {"$first": "$tracks.name"},
        "album": {"$first": "$name"},
        "popularity": {"$first": "$tracks.popularity"}
    }},
    {"$sort": {"popularity": -1}}
]

# mydoc = albums.aggregate(pipeline)
# for doc in mydoc:
#    pprint.pprint(doc)

# measure the execution time
for i in range(30):
    start_time = time.time()
    albums.aggregate(pipeline)
    mongo_query_times["popular_featured_track_query"].append(time.time() - start_time)
```

### Update

```python
from pymongo import MongoClient
import time

mongo_query_times["update"] = []

client = MongoClient('localhost', 27017)
db = client.spotify

artists = db.artists
tracks = db.tracks

# update the artists with name "dan berk" to "new_artist_name"
for i in range(30):
    start_time = time.time()
    artists.update_one({"name": "dan berk"}, {"$set": {"name": "new_artist_name"}})
    tracks.update_many({"artists.name": "dan berk"}, {"$set": {"artists.$.name": "new_artist_name"}})
    mongo_query_times["update"].append(time.time() - start_time)
```

### Insert

```python
from pymongo import MongoClient
import time

client = MongoClient('localhost', 27017)
db = client.spotify

mongo_query_times["insert"] = []

artists = db.artists
tracks = db.tracks
albums = db.albums

for i in range(30):
    start = time.time()
    # Insert an artist with name "inserted_artist"
    artist_data = {"name": f"inserted_artist{i}"}
    artist_id = artists.insert_one(artist_data).inserted_id

    # Insert a track of this album with name "inserted_track" and other attributes
    track_data = {
        "name": "inserted_track{i}",
        "artists": [{"_id": artist_id, "name": "inserted_artist{i}"}],
        "popularity": 100,
        "duration_ms": 5000,
        "explicit": 1,
        "loudness": -5
    }
    track_id = tracks.insert_one(track_data).inserted_id

    # Insert an album of this artist with name "inserted_album"
    album_data = {"name": "inserted_album{i}", "artist_id": artist_id,"tracks": [track_data]}
    album_id = albums.insert_one(album_data).inserted_id
    mongo_query_times["insert"].append(time.time() - start)
```
